```python
import json,numpy as np,os,time
from typing import Dict,List,Tuple,Optional,Callable
from datetime import datetime
from dataclasses import dataclass,field
from collections import Counter,defaultdict
print(f"üêã v5-Lite | {datetime.now().strftime('%H:%M:%S')}")
class C:
 K=os.path.exists('/kaggle/input')
 I='/kaggle/input/arc-prize-2025/arc-agi_test_challenges.json'if K else'data/arc-agi_test_challenges.json'
 O='/kaggle/working/submission.json'if K else'submission.json'
 L='/kaggle/working/log.txt'if K else'log.txt'
c=C()
@dataclass
class SS:
 n:str;t:int=0;i:List[str]=field(default_factory=list)
class PT:
 def __init__(s):s.d={};s.tt=0;s.t0=time.time()
 def r(s,n,i):
  if n not in s.d:s.d[n]=SS(n=n)
  s.d[n].t+=1
  if i not in s.d[n].i:s.d[n].i.append(i)
 def st(s,t):s.tt=t
 def c(s,n):return s.d[n].t/s.tt if n in s.d and s.tt else 0
 def g(s):{k:{'t':v.t,'c':s.c(k)}for k,v in s.d.items()}
pt=PT()
class L:
 def __init__(s,p):
  s.p=p;s.t0=time.time()
  with open(s.p,'w')as f:f.write(f"v5-Lite {datetime.now()}\n")
 def l(s,m,co=True):
  with open(s.p,'a')as f:f.write(f"[{time.time()-s.t0:.1f}s] {m}\n")
  if co:print(m)
lg=L(c.L)
lg.l('Init OK')
```

```python
class V:
 @staticmethod
 def vg(g,ctx=''):
  if not isinstance(g,list)or not g:return False,'Empty'
  if not all(isinstance(r,list)for r in g):return False,'Not 2D'
  if g and not all(len(r)==len(g[0])for r in g):return False,'Ragged'
  for i,r in enumerate(g):
   for j,v in enumerate(r):
    if not isinstance(v,(int,np.integer))or v<0 or v>9:return False,f'[{i}][{j}] bad'
  return True,'OK'
 @staticmethod
 def vs(s,tc):
  lg.l('\n'+'='*70);lg.l('VALIDATING');lg.l('='*70)
  if not isinstance(s,dict):return False,'Not dict'
  lg.l('‚úÖ Dict OK')
  if set(tc.keys())-set(s.keys()):return False,'Missing tasks'
  lg.l(f'‚úÖ {len(s)} tasks')
  for tid,preds in s.items():
   if not isinstance(preds,list):return False,f'{tid}: not list'
   for i,p in enumerate(preds):
    if 'attempt_1'not in p or'attempt_2'not in p:return False,f'{tid}[{i}]: no attempts'
    for a in['attempt_1','attempt_2']:
     ok,_=V.vg(p[a],f'{tid}[{i}][{a}]')
     if not ok:return False,_
  lg.l('‚úÖ Valid');lg.l('='*70+'\n')
  return True,'OK'
lg.l('Validator OK')
```

```python
class SD:
 @staticmethod
 def ci(td,i=0):return td['test'][i]['input']if'test'in td else[[0]]
 @staticmethod
 def cl(g):return g.tolist()if isinstance(g,np.ndarray)else g
lg.l('Defaults OK')
```

```python
def gt(g):return tuple(tuple(r)for r in g)
def ge(g1,g2):return np.array_equal(np.array(g1),np.array(g2))
def sh(g):return(len(g),len(g[0])if g and g[0]else 0)if g else(0,0)
def fl(g):return[v for r in g for v in r]
@dataclass
class DO:
 id:int;c:int;px:List[Tuple[int,int]];bb:Tuple[int,int,int,int]
 @property
 def a(s):return len(s.px)
def fcc(g,bg=0):
 a=np.array(g);h,w=a.shape;v=np.zeros((h,w),dtype=bool);o=[];oid=0
 for i in range(h):
  for j in range(w):
   if v[i,j]or a[i,j]==bg:continue
   co=a[i,j];px=[];st=[(i,j)]
   while st:
    r,c=st.pop()
    if r<0 or r>=h or c<0 or c>=w or v[r,c]or a[r,c]!=co:continue
    v[r,c]=True;px.append((r,c));st.extend([(r-1,c),(r+1,c),(r,c-1),(r,c+1)])
   if px:
    rs=[p[0]for p in px];cs=[p[1]for p in px]
    o.append(DO(oid,co,px,(min(rs),min(cs),max(rs),max(cs))));oid+=1
 return o
def docc(td):
 tp=td.get('train',[])
 if not tp:return None
 try:
  cm=[]
  for p in tp:
   inp,out=p['input'],p['output']
   if np.array(inp).shape!=np.array(out).shape:return None
   io=fcc(inp);oo=fcc(out)
   if len(io)!=len(oo):return None
   pm={}
   for iobj,oobj in zip(io,oo):
    if sorted(iobj.px)!=sorted(oobj.px):return None
    if iobj.c in pm and pm[iobj.c]!=oobj.c:return None
    pm[iobj.c]=oobj.c
   cm.append(pm)
  if not cm:return None
  fm=cm[0]
  if not all(m==fm for m in cm)or all(k==v for k,v in fm.items()):return None
  return{'t':'occ','m':fm,'cf':0.85}
 except:return None
def aocc(ti,m):
 a=np.array(ti);r=a.copy()
 for o,n in m.items():r[a==o]=n
 return r.tolist()
def dcmr(tp):
 if not tp:return None
 try:
  inp,out=tp[0]['input'],tp[0]['output']
  if sh(inp)!=sh(out):return None
  m={}
  for iv,ov in zip(fl(inp),fl(out)):
   if iv in m and m[iv]!=ov:return None
   m[iv]=ov
  if all(k==v for k,v in m.items()):return None
  for p in tp[1:]:
   if sh(p['input'])!=sh(p['output']):return None
   for iv,ov in zip(fl(p['input']),fl(p['output'])):
    if m.get(iv,iv)!=ov:return None
  return{'t':'cm','m':m,'cf':0.90}
 except:return None
def acmr(g,m):return[[m.get(v,v)for v in r]for r in g]
def dstr(tp):
 if not tp:return None
 try:
  rt=[]
  for p in tp:
   ih,iw=sh(p['input']);oh,ow=sh(p['output'])
   if ih==0 or iw==0:return None
   rt.append((oh/ih,ow/iw))
  if not all(r==rt[0]for r in rt):return None
  hr,wr=rt[0]
  if hr!=int(hr)or wr!=int(wr)or(hr==1 and wr==1):return None
  hs,ws=int(hr),int(wr)
  for p in tp:
   if not np.array_equal(np.tile(np.array(p['input']),(hs,ws)),np.array(p['output'])):return None
  return{'t':'st','hs':hs,'ws':ws,'cf':0.95}
 except:return None
def astr(g,hs,ws):return np.tile(np.array(g),(hs,ws)).tolist()
def dri(td):
 tp=td.get('train',[])
 if not tp:return None
 r=dcmr(tp)
 if r:return('rcm',lambda g:acmr(g,r['m']),r['cf'])
 r=dstr(tp)
 if r:return('rt',lambda g:astr(g,r['hs'],r['ws']),r['cf'])
 return None
def r90(g):return np.rot90(np.array(g),k=-1).tolist()
def fh(g):return np.fliplr(np.array(g)).tolist()
def fv(g):return np.flipud(np.array(g)).tolist()
def dcm(ig,og):
 inp,out=np.array(ig),np.array(og)
 if inp.shape!=out.shape:return None
 m={}
 for iv,ov in zip(inp.flatten(),out.flatten()):
  if iv in m and m[iv]!=ov:return None
  m[iv]=ov
 return m if not all(k==v for k,v in m.items())else None
def acm(g,m):
 a=np.array(g);r=a.copy()
 for o,n in m.items():r[a==o]=n
 return r.tolist()
def dpm(td):
 tp=td.get('train',[])
 if not tp:return None
 tr=[('r90',r90),('r180',lambda g:r90(r90(g))),('fh',fh),('fv',fv)]
 for n,tf in tr:
  if all(ge(tf(p['input']),p['output'])for p in tp):return(n,tf)
 fm=dcm(tp[0]['input'],tp[0]['output'])
 if fm and all(dcm(p['input'],p['output'])==fm for p in tp):return('cm',lambda g:acm(g,fm))
 return None
def dsy(g):
 a=np.array(g);h,w=a.shape
 l,r=a[:,:w//2],np.fliplr(a[:,w//2:])
 if l.shape==r.shape and 0.6<=np.mean(l==r)<0.95:return'h'
 t,b=a[:h//2,:],np.flipud(a[h//2:,:])
 if t.shape==b.shape and 0.6<=np.mean(t==b)<0.95:return'v'
 return None
def asy(g,st):
 a=np.array(g);h,w=a.shape
 if st=='h':
  l=a[:,:w//2];return np.hstack([l,np.fliplr(l)]).tolist()
 elif st=='v':
  t=a[:h//2,:];return np.vstack([t,np.flipud(t)]).tolist()
 return g
@dataclass
class SP:
 g:List[List[int]];cf:float;n:str
def cp(ti,td,tid):
 p=[]
 try:
  r=dri(td)
  if r:
   n,tf,cf=r;pr=SD.cl(tf(ti));p.append(SP(pr,cf,n));pt.r(n,tid)
 except:pass
 try:
  r=dpm(td)
  if r:
   n,tf=r;pr=SD.cl(tf(ti));p.append(SP(pr,0.85,f'p{n}'));pt.r(f'p{n}',tid)
 except:pass
 try:
  r=docc(td)
  if r and r['t']=='occ':
   pr=aocc(ti,r['m']);p.append(SP(pr,r['cf'],'occ'));pt.r('occ',tid)
 except:pass
 try:
  st=dsy(ti)
  if st:
   pr=SD.cl(asy(ti,st));p.append(SP(pr,0.75,f'sy{st}'));pt.r(f'sy{st}',tid)
 except:pass
 return p
def vp(p,at,ti):
 if not p:return ti
 vg={}
 for pr in p:
  gk=gt(pr.g)
  if gk not in vg:vg[gk]={'g':pr.g,'tc':0,'s':[]}
  vg[gk]['tc']+=pr.cf;vg[gk]['s'].append(pr.n)
 rk=sorted(vg.values(),key=lambda x:x['tc'],reverse=True)
 return rk[0]['g']if at==1 else(rk[0]['g']if rk[0]['tc']>0.95 else(rk[1]['g']if len(rk)>1 else ti))
def es(ti,td,at=1,tid=''):
 try:
  p=cp(ti,td,tid)
  if p:lg.l(f"  {tid}: {len(p)} ‚Üí {[x.n for x in p]}",co=False)
  r=vp(p,at,ti);return SD.cl(r)
 except:return ti
lg.l('Solver OK')
```

```python
def gs(tc):
 lg.l('\n'+'='*70);lg.l('GENERATING');lg.l('='*70)
 s={};tt=len(tc);pt.st(tt);t0=time.time()
 for ix,(tid,td)in enumerate(tc.items(),1):
  if ix%50==0:
   el=time.time()-t0;rt=ix/el if el>0 else 0;rm=(tt-ix)/rt if rt>0 else 0
   lg.l(f"  {ix}/{tt} ({ix/tt*100:.1f}%) | {rt:.1f}/s | ETA {rm/60:.1f}m")
  tp=[]
  for tix in range(len(td['test'])):
   ti=td['test'][tix]['input']
   a1=es(ti,td,1,tid);a2=es(ti,td,2,tid)
   tp.append({'attempt_1':SD.cl(a1),'attempt_2':SD.cl(a2)})
  s[tid]=tp
 el=time.time()-t0;lg.l(f"\n‚úÖ {tt} tasks in {el:.1f}s");lg.l('='*70+'\n')
 lg.l('\n'+'='*70);lg.l('STATS');lg.l('='*70)
 st=pt.g()
 if st:
  lg.l(f"  {'Solver':<20} {'Trig':>8} {'Cov%':>8}")
  for n,d in sorted(st.items(),key=lambda x:x[1]['c'],reverse=True):
   lg.l(f"  {n:<20} {d['t']:>8} {d['c']*100:>7.1f}")
 lg.l('='*70+'\n')
 return s
lg.l('Gen OK')
```
